## 5.1. 리스트 다루기

```python
# 리스트 변수 할당

nums = [1, 2, 3, 4, 5]
cars = ['taxi', 'bus', 'truck']
money = ['1000원', '1dollar', '20엔', 10000]

print(nums)
line()
print(cars)
line()
print(money)
```

```python
# 리스트 인덱싱

money = ['1000원', '1dollar', '20엔', 10000]

print(money[0])
line()
print(money[1])
line()
print(money[-1])

```

```python
# 리스트 슬라이싱

money = ['1000원', '1dollar', '20엔', 10000]

print(money[:1])
line()
print(money[1:])
line()
print(money[1:3])

```

```python
# 리스트 원소 분리
# 이게 가능하려면, 리스트의 원소 개수만큼 변수를 할당해 줘야 함.

m1, m2, m3, m4 = won

print(m1)
line()
print(m4)
```

```python
# 리스트 원소 변경

cheeses = ['체다', '모짜렐라', '리코타']

print(cheeses)
line()

cheeses[0] = '까망베르'
print(cheeses)
```

```python
# 리스트 더하기

cheeses = ['체다', '모짜렐라', '리코타']
cheese = ['까망베르']

print(cheeses)
line()
print(cheese)
line()

new_cheeses = cheeses + cheese
print(new_cheeses)
```

```python
# 리스트 메서드1
# index 메서드: 리스트 요소의 index를 찾아주는 메서드

cheeses = ['체다', '모짜렐라', '리코타', '까망베르']

print(cheeses)
line()
print(cheeses.index('체다'))
```

```python
# 리스트 메서드2
# index 메서드: 리스트 요소의 index를 찾아주는 메서드

cheeses = ['체다', '모짜렐라', '리코타', '까망베르']

print(cheeses)
line()
cheeses.append('고르곤')
print(cheeses)
```

```python
# 리스트 메서드3
# insert 메서드: 리스트에서 원하는 index에 요소를 끼워 넣는 메서드

cheeses = ['체다', '모짜렐라', '리코타', '까망베르']

print(cheeses)
line()
cheeses.insert(1, '고르곤')
print(cheeses)
```

```python
# 리스트 메서드4
# extend 메서드: 본 리스트에 둘 이상의 요소로 구성된 리스트를 연장시키는 메서드
# 두 리스트를 더하는 효과(+)와 동일함.

cheeses = ['체다', '모짜렐라', '리코타', '까망베르']

print(cheeses)
line()
cheeses.extend(['피자', '핫도그'])
print(cheeses)
```

```python
# 리스트 메서드5
# sort 메서드 : 정렬
# 원본 변경에 주의

nums = [200, 500, 100, 1000]

print(nums)
line()
nums.sort()
print(nums)
```

```python
# 리스트 메서드6
# reverse 메서드 : 역순 정렬 -> 이건 크기에 따른 정렬이 아니고, 위치 역순 정렬임.
# 원본 변경에 주의

nums = [200, 500, 100, 1000]

print(nums)
line()
nums.reverse()
print(nums)
```

```python
# 리스트 메서드7_1
# pop 메서드 : 리스트의 마지막 요소를 return으로 넘겨주고, 원본 리스트에서는 제거하는 메서드
# 원본 변경에 주의

nums = [200, 500, 100, 1000]

print(nums)
line()
return_val = nums.pop()
print(return_val)
line()
print(nums)
```

```python
# 리스트 메서드7_1
# pop 메서드 : 리스트의 마지막 요소를 return으로 넘겨주고, 원본 리스트에서는 제거하는 메서드
# 메서드 인자로 index를 넣어주면 해당 index의 요소가 제거되고, return 됨.

nums = [200, 500, 100, 1000]

print(nums)
line()
return_val = nums.pop(1)
print(return_val)
line()
print(nums)
```

```python
# 리스트 메서드8
# remove 메서드 : 메서드의 인자로 리스트의 요소를 넣어주면, 해당 요소가 제거됨.
# index 정보로 요소를 제거할 때는 pop, 요소 정보로 제거할 때는 remove를 사용함.
# 원본 변경에 주의

nums = [200, 500, 100, 1000]

print(nums)
line()
nums.remove(500)
print(nums)
```

```python
# 리스트 메서드8
# count 메서드 : 메서드의 인자로 리스트의 요소로 넣어주고, 해당 요소가 리스트에서 몇 번 등장하는지 알려주는 메서드

nums = [200, 500, 100, 1000, 500]

print(nums)
line()
print(nums.count(500))
```

```python
# 파이썬 내장함수와 리스트
# 리스트와 궁합이 잘맞는(그만큼 자주 사용되는) 파이썬 내장함수들

nums = [200, 500, 100, 1000, 500]

print(nums)
line()
print(len(nums))
line()
print(max(nums))
line()
print(min(nums))
line()
print(sum(nums))
line()
print(sorted(nums))

```

# 7. 딕셔너리
* '사전'이란 의미인데, key(색인)와 value(값) 세트로 된 구조
* 마치 사전에 수록된 모양으로 된 데이터 형식과 비슷함.
* **낱말**과 **뜻풀이** 형태로 된 형식
* 딕셔너리에서는 이를 각각 **키(key)**와 **값(value)**이라고 부릅니다.

## 7.1. 딕셔너리 다루기

```python
# dict 변수 할당1
# base

dic = {'name':'serendipi27', 'age': 1000, 'color':'red'}

print(dic)
line()
print(type(dic))
```

```python
# dict 변수 할당2
# dict() 함수 이용 + tuple(원소 두개로 이루어진 리스트 또는 tuple)
# 이중 리스트나 이중 튜플 모두 다 되고, 내부 리스트에는 2개의 요소가 들어가 있으면 됨.

tup = ((1, '서울'), (2, '대전'), (3, '대구'), (4, '부산'))

dic = dict(tup)

print(dic)
line()
print(type(dic))
```

```python
# dict 변수 할당3
# 리스트 컴프레헨션을 이용한 dict 변수 할당
# 외부 리스트든, 외부 튜플이든 모두 동일한 결과를 얻음.

dic1 = dict((i, i**2) for i in range(5))  # 튜플은 양 괄호(())를 생략 가능하기 때문에 이 코드는 True임.
dic2 = dict([(i, i**2) for i in range(5)])

print(dic1)
print(type(dic1))
line()
print(dic2)
print(type(dic2))
```

```python
# dict의 key를 이용한 value 찾기

tup = ((1, '서울'), (2, '대전'), (3, '대구'), (4, '부산'))

dic = dict(tup)

print(dic[1])
line()
print(dic.get(1))
line()
line()

print(dic[4])
line()
print(dic.get(4))
line()
```

```python
# dict의 value 수정하기

print(dic)
line()

dic[1] = '일산'
print(dic)
```

```python
# dict의 새로운 value 추가하기
# DataFrame에서 새로운 열을 만들고 Series를 추가하는 것과 비슷함.
# key-value의 dict base임.

print(dic)
line()

dic[10] = '하동'
print(dic)
```

```python
# dict 메서드
# items(), keys(), values()
# 실제 사용을 하려면, 이 메서드 적용 후, list(), tuple()로 wrapping 해줘야 함.
# 또는 iterator로 활용할 것.

print(dic)
line()

print(dic.keys())
line()
print(dic.values())
line()
print(dic.items())
line()
```

# 8. 셋
* 수학에 나오는 집합(set) 연산을 처리하기 위한 자료형
* 중복을 허용하지 않음
* 순서를 고려하지 않음
  * 순서를 고려해 처리해야 한다면 리스트나 튜플 자료형으로 변환이 필요함.

```python
# set은 리스트를 제외한 모든 데이터형을 요소로 가질 수 있음.

s = {'hi', 0, 1.5, True, (1,2,3)}

print(s)
```

```python
# 리스트는 셋의 요소가 될 수 없음.
# 셋은 중복을 허용하지 않으므로, 가변 객체를 가질 수 없게 되어 있음.

s = {"hi", 0, 1, [10, 20, 30], 'good', 1, 2, 0}
```

```python
# 셋은 중복되는 요소는 하나씩만 남기고 중복 제거함.

s = {"hi", 0, 1, 'good', 1, 2, 0}

print(s)
```

```python
# 집합은 순서가 없기 때문에 인덱싱, 슬라이싱이 불가능함.

s[0]
```

```python
# 집합 메서드1
# add() 메서드: 집합에 요소 추가하기

fav = {'sky', 'air'}

print(fav)
line()
print(type(fav))
line()

fav.add('sea')
print(fav)
```

```python
# 집합 메서드2
# update() 메서드: 여러 요소 한꺼번에 추가하기

print(fav)
line()

fav.update('soil', 'river', {0, 1})
print(fav)
```

```python
# 집합 메서드3
# discard() 메서드: 특정 요소 삭제. 하나씩이며 튜플이나 셋 형태로 여러 개 삭제 불가능

print(fav)
line()

fav.discard('i')
print(fav)
line()
```

```python
print(fav)
line()
fav.discard('r')
print(fav)
```

```python
# 집합 메서드4
# pop() 메서드: 리스트와 비슷함. 그러나 주의해야할 것은 random하게 요소가 추출됨.
# 리스트의 pop과는 다르게 제거하고자 하는 value의 index를 인자로 넣을 수 없음. 아무런 인자를 넣을 수 없음.
# random 요소가 추출된다고 하는데, 경험상 set의 가장 앞부분의 요소가 나오는 것 같고.
# 그렇지만 set은 순서가 없기 때문에 앞뒤를 상정하는 것 자체가 어려운 개념임.

print(fav)
line()

print(fav.pop())
print(fav)
line()
```

```python
# 집합 연산1
# 합(|), 교(&), 차(-), 대칭차(^)

sa = {1,2,3,4,5,6}
sb = {1,2,3,6,7,8}

print(sa | sb)
line()
print(sa & sb)
line()
print(sa - sb)
line()
print(sa ^ sb)

```

```python
# 부분집합 확인
# <=, >=

sa = {1,2,3,4,5,6}
sb = {1,2,3,6,7,8}

print({4,5,6} <= sa)
print({4,5,6} >= sa)
line()
print({4,5,6} <= sb)
print({4,5,6} >= sb)
```

```python
# 집합 관련 함수
# len, max, min, max, sum, sorted, 멤버십 테스트

nums = {1, 2, 3, 3.14, 10, 200, 5000}

print(1 in nums)
line()
print(1 not in nums)
line()

```

## 11.1 변수 출력 다루기

```python
# %를 이용하는 방법

print('%s' %('hello'))
line()
print('%d' %20)
line()
print('%f' %0.023)
line()
print('%.2f' %0.023)
```

```python

name = 'jason'
age = 1200

print('hello, 내 이름은 %s이고, 나이는 %d살이야.' %(name, age))

```

```python
# format을 이용하는 방법

name = 'jason'
age = 1200.0005

print('{}'.format('hello'))
line()
print('hello, 내 이름은 {}이고, 나이는 {}살이야.'.format(name, age))
line()
print('hello, 내 이름은 {}이고, 나이는 {:.1f}살이야.'.format(name, age))


```

```python
# f-string을 이용하는 방법

name = 'jason'
age = 1200.0005

print(f'내 이름은 {name}이고, 나이는 {age:.1f}야. 넌??')
```
